# 机器人

```python
#从刚开始的第31点开始

def trade_start_stop(data,i):
    avg_5 = sum(data['4. close'][i-5:i])/5
    avg_10 = sum(data['4. close'][i-10:i])/10
    avg_20 = sum(data['4. close'][i-20:i])/20
    avg_30 = sum(data['4. close'][i-30:i])/30
    if avg_5 > avg_10 and avg_10 > avg_20 and avg_20 > avg_30:
        return "All in"#多头排列
    elif avg_5 < avg_10 and avg_10 > avg_20 and avg_20 > avg_30:
        return "All sell out" #多头排列后期，出现5日回落死叉
    elif avg_5 < avg_10 and avg_10 < avg_20 and avg_20 < avg_30:
        return "Empty"#空头排列
    elif avg_5 < avg_10 and avg_10 > avg_20 and avg_20 < avg_30:
        return "All in #空头排列后期"
    else:
        return "No action"

output = []   
for i in range(30,len(data)):
    output.append(trade_start_stop(data,i))
```

# 根据机器人判断展现盈亏情况

```python
# Money指代本金，
def Money_Change(data,Money,Money_Change):
    for i in range(len(output)):
        if output[i] == 'All in' or output[i] == "All in #空头排列后期":
            if i + 1 < len(output):
                Money = Money*(data['4. close'][i+1]/data['4. close'][i])
                Money_Change.append(Money)
        else:
            Money_Change.append(Money)
    return Money_Change
        
```

# 实际操作机器人

```python
class solution(object):
    def __init__(self, total,output,ans):
        self.total = total
        self.output = output
        self.ans = ans
    def trade_operation(self):
        for i in range(len(self.output)):

            if self.output[i] == "All in" and self.total == 1:
                self.ans.append('多头排列，但我们已经满仓，无需加仓。')
            elif self.output[i] == "All in" and self.total == 0:
                self.total = 1
                self.ans.append("多头排列，All in 加仓。")
            elif self.output[i] == "All sell out" and self.total == 1:
                self.total = 0
                self.ans.append("多头排列后期，出现5日回落死叉，急需清仓！")
            elif self.output[i] == "All sell out" and self.total == 0:
                self.ans.append("多头排列后期，出现5日回落死叉，但我们已经空仓，无需清仓。")
            elif self.output[i] == "Empty" and self.total == 1:
                self.total = 0
                self.ans.append("空头排列,急需清仓!")
            elif self.output[i] == "Empty" and self.total == 0:
                self.ans.append("空头排列,但我们已经空仓，无需清仓。")
            elif self.output[i] == "All in #空头排列后期" and self.total == 1:
                self.ans.append("空头排列后期,但我们已经满仓，无需加仓。")
            elif self.output[i] == "All in #空头排列后期" and self.total == 0:
                self.total = 1
                self.ans.append("空头排列后期,All in 加仓")
            elif self.output[i] == "No action":
                self.ans.append("No action")
            else:
                self.ans.append("异常情况")
        if len(self.ans) == len(self.output):
            return self.ans

solution(0,output,[]).trade_operation()
    
```

## 实际操作下的买入卖出盈亏比

```python
Operatation = solution(0,output,[]).trade_operation()
```

```python
# 计算什么卖出的时候钱和之前一次买入的时候比，是增加了还是减少了。增加了则表示赚钱，减少了则表示亏钱
Money_add_lose = []#标记状态，第一个肯定是all in, 第二个才是清仓，因此有规律
for i in range(len(Operatation)):
    if Operatation[i] == '多头排列，All in 加仓。' or Operatation[i] == '空头排列后期,All in 加仓':
        Money_add_lose.append(i) 
    if Operatation[i] == "多头排列后期，出现5日回落死叉，急需清仓！" or Operatation[i] == "空头排列,急需清仓!":
        Money_add_lose.append(i)
```

```python
Money_status = [] # 表示盈亏情况
for i in range(0,len(Money_add_lose),2):
    if i + 1 < len(Money_add_lose):
        if data['4. close'][Money_add_lose[i+1]]/data['4. close'][Money_add_lose[i]] >= 1:
            Money_status.append(1)
        else:
            Money_status.append(0)
```

```python
# 盈利比亏损
sum(Money_status)/len(Money_status)
```
